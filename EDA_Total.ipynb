# Análisis Exploratorio de Datos de los tres Sets
## Carga de Datos

```python
import numpy as np 
import pandas as pd 
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv("combined_csv_without_features.csv")

df
```

## Análisis Exploratorio de Datos
### Determinando el tamaño de muestreo

```python
import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns

for n in [10, 50, 100, 500, 1000, 5000, 10000, 50000]:
    
    df1 = df[df['actividad'] == 'Standing'].head(n)
    df2 = df[df['actividad'] == 'Walking'].head(n)
    df3 = df[df['actividad'] == 'Stairs'].head(n)
    df4 = df1.append(df2).append(df3)
    
    plt.figure(figsize=(15,28*4))
    gs = gridspec.GridSpec(28, 1)
    
    for i, cn in enumerate(df4[['gFx','gFy','gFz','TgF']]):
        ax = plt.subplot(gs[i])
        sns.distplot(df4[cn][df4.actividad == 'Standing'])
        sns.distplot(df4[cn][df4.actividad == 'Walking'])
        sns.distplot(df4[cn][df4.actividad == 'Stairs'])
        ax.set_xlabel('')
        plt.legend(['Upstairs', 'Standing', 'Walking'])
        ax.set_title('histogram for ' + str(cn) + ' with size of ' + str(n))
    
    plt.show()
```
