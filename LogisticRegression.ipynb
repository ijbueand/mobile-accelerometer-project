```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('combined_and_transformed_data.csv')

df
```

```python
# Read train and test datasets
df_train = pd.read_csv('train_data.csv')
df_test = pd.read_csv('test_data.csv')
```

```python
# Select most-correlated features
df_train.corr()['actividad_cat'].sort_values(ascending = False, key=np.abs)
```

```python
# Split the data into X and y sets
y_test = df_test['actividad_cat']
y_train = df_train['actividad_cat']

X_test = df_test.drop(columns={'actividad_cat', 'sujeto', 'actividad', 'gFz_mean', 'movil', 'userID', 'TgF_kurtosis', 'gFy_kurtosis', 'gFz_median'}, axis = 1)
X_train = df_train.drop(columns={'actividad_cat', 'sujeto', 'actividad', 'gFz_mean', 'movil', 'userID', 'TgF_kurtosis', 'gFy_kurtosis', 'gFz_median'}, axis = 1)

X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, f1_score

# Fit and train the logistic regression model
log_reg = LogisticRegression(max_iter=500, 
                             random_state=0, 
                             penalty='none',
                             solver = 'lbfgs',
                             class_weight='balanced')

log_reg.fit(X_train, y_train)

# Predict the classes using the test set
y_pred = log_reg.predict(X_test)

# Generate the confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix\n", confusion_mat)

# Calculate F1-score
f1 = f1_score(y_test, y_pred, average='weighted')
print("\n F1 score : %.2f" % f1)
```

```python
# Guardamos el modelo en formato .sav
import pickle

pickle.dump(log_reg, open("logistic_regression_model.sav", "wb"))
```
