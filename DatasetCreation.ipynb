# Elaboración del dataset integrado

```python
import os
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

column_list = ['time', 'gFx', 'gFy', 'gFz', 'TgF', 'actividad', 'movil', 'sujeto']
combined_csv = pd.DataFrame()

csv_files = [f for f in os.listdir('all_data') if f.endswith('.csv')]
for file in csv_files:
    df = pd.read_csv('all_data\\' + file, delimiter=";", decimal=",")
    df.columns = column_list
    print('Se ha cargado', file)
    duration = df['time'].iloc[-1]
    df = df[(df['time'] > 1) & (df['time'] < (duration - 1))]
    combined_csv = combined_csv.append(df, ignore_index=True)
    
combined_csv
```

```python
for col in ['actividad', 'movil', 'sujeto']:
        combined_csv[col] = combined_csv[col].str.capitalize().str.strip()
        
combined_csv
```

```python
combined_csv.to_csv("combined_csv_without_features.csv", index=False, header=True)
```
